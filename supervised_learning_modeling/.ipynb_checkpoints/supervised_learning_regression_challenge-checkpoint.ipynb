### Feature Engineering

```python
# numerical features I'm interested in
house_features = house_prices[['garagecars', 'overallqual', 
                              'lotarea', 'totalbsmtsf', 
                               'grlivarea', 'secondflrsf']].copy()
```

```python
# consolidate multiple features
house_features['tot_sq_ft'] = house_features.totalbsmtsf + \
                              house_features.grlivarea + \
                              house_features.secondflrsf
# remove extra columns
house_features.drop(['totalbsmtsf', 'grlivarea', 'secondflrsf'], axis=1, inplace=True)
```

```python
# examine contiuous features for outliers
fig = plt.figure(figsize=(15, 3))

for i, column in enumerate(list(house_features.columns)):
    plt.subplot(ceil(len(list(house_features.columns))/5), 5, i+1)
    plt.title(column)
    plt.boxplot(house_features[column])
    plt.xticks([], [])
```

```python
# winsorize to remove outliers
for column in (list(house_features.columns)):
    house_features[column] = winsorize(house_features[column], (.01, 0.05))
```

```python
# examine to make sure outliers have been removed
# everything looks good
fig = plt.figure(figsize=(15, 3))

for i, column in enumerate(list(house_features.columns)):
    plt.subplot(ceil(len(list(house_features.columns))/5), 5, i+1)
    plt.title(column)
    plt.boxplot(house_features[column])
    plt.xticks([], [])
```

```python
# combine continuous and categorical features
house_features = pd.concat((house_features,
         pd.get_dummies(house_prices.neighborhood),
         pd.get_dummies(house_prices.exterior1st),
         pd.get_dummies(house_prices.exterior2nd)), axis=1)
```

```python
house_features.head()
```

```python
house_features.shape
```

```python
# assign X and Y variables

# Y is the target variable
Y = house_prices.saleprice
# X is the feature set
X = house_features
```

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 465)

print("The number of observations in training set is {}".format(X_train.shape[0]))
print("The number of observations in test set is {}".format(X_test.shape[0]))
```

<a id="linear"></a>
### Linear Regression

```python
# We fit an OLS model using sklearn
lrm = LinearRegression()
lrm.fit(X_train, y_train)

# We are making predictions here
y_preds_train = lrm.predict(X_train)
y_preds_test = lrm.predict(X_test)

print("R-squared of the model in the training set is: {}".format(lrm.score(X_train, y_train)))
print("-----Test set statistics-----")
print("R-squared of the model in the test set is: {}".format(lrm.score(X_test, y_test)))
print("Mean absolute error of the prediction is: {}".format(mean_absolute_error(y_test, y_preds_test)))
print("Mean squared error of the prediction is: {}".format(mse(y_test, y_preds_test)))
print("Root mean squared error of the prediction is: {}".format(rmse(y_test, y_preds_test)))
print("Mean absolute percentage error of the prediction is: {}".format(np.mean(np.abs((y_test - y_preds_test) / y_test)) * 100))
```

```python
alphas = [np.power(10.0,p) for p in np.arange(-10,40,1)]
```

<a id="ridge"></a>
### Ridge Regression

```python
# Fitting a ridge regression model. Alpha is the regularization
# parameter (usually called lambda). As alpha gets larger, parameter
# shrinkage grows more pronounced.
ridge_cv = RidgeCV(alphas=alphas, cv=5) 
ridge_cv.fit(X_train, y_train)

# We are making predictions here
y_preds_train = ridge_cv.predict(X_train)
y_preds_test = ridge_cv.predict(X_test)

print("R-squared of the model on the training set is: {}".format(ridge_cv.score(X_train, y_train)))
print("-----Test set statistics-----")
print("R-squared of the model on the test set is: {}".format(ridge_cv.score(X_test, y_test)))
print("Mean absolute error of the prediction is: {}".format(mean_absolute_error(y_test, y_preds_test)))
print("Mean squared error of the prediction is: {}".format(mse(y_test, y_preds_test)))
print("Root mean squared error of the prediction is: {}".format(rmse(y_test, y_preds_test)))
print("Mean absolute percentage error of the prediction is: {}".format(np.mean(np.abs((y_test - y_preds_test) / y_test)) * 100))
```

<a id="lasso"></a>
### Lasso Regression

```python
lasso_cv = LassoCV(alphas=alphas, cv=5)

lasso_cv.fit(X_train, y_train)

# We are making predictions here
y_preds_train = lasso_cv.predict(X_train)
y_preds_test = lasso_cv.predict(X_test)

print("Best alpha value is: {}".format(lasso_cv.alpha_))
print("R-squared of the model in training set is: {}".format(lasso_cv.score(X_train, y_train)))
print("-----Test set statistics-----")
print("R-squared of the model in test set is: {}".format(lasso_cv.score(X_test, y_test)))
print("Mean absolute error of the prediction is: {}".format(mean_absolute_error(y_test, y_preds_test)))
print("Mean squared error of the prediction is: {}".format(mse(y_test, y_preds_test)))
print("Root mean squared error of the prediction is: {}".format(rmse(y_test, y_preds_test)))
print("Mean absolute percentage error of the prediction is: {}".format(np.mean(np.abs((y_test - y_preds_test) / y_test)) * 100))
```

<a id="elastic"></a>
### ElasticNet Regression

```python
elasticnet_cv = ElasticNetCV(alphas=alphas, cv=5)

elasticnet_cv.fit(X_train, y_train)

# We are making predictions here
y_preds_train = elasticnet_cv.predict(X_train)
y_preds_test = elasticnet_cv.predict(X_test)

print("Best alpha value is: {}".format(elasticnet_cv.alpha_))
print("R-squared of the model in training set is: {}".format(elasticnet_cv.score(X_train, y_train)))
print("-----Test set statistics-----")
print("R-squared of the model in test set is: {}".format(elasticnet_cv.score(X_test, y_test)))
print("Mean absolute error of the prediction is: {}".format(mean_absolute_error(y_test, y_preds_test)))
print("Mean squared error of the prediction is: {}".format(mse(y_test, y_preds_test)))
print("Root mean squared error of the prediction is: {}".format(rmse(y_test, y_preds_test)))
print("Mean absolute percentage error of the prediction is: {}".format(np.mean(np.abs((y_test - y_preds_test) / y_test)) * 100))
```

<a id="visualization"></a>
### Visualize Differences

```python
# combine scores for each model into lists
model = [lrm,
        ridge_cv,
        lasso_cv,
        elasticnet_cv]

training = []
test = []

for x in model:
    training.append(x.score(X_train, y_train))
    test.append(x.score(X_test, y_test))
```

```python
# combine lists into a dataframe
df = pd.DataFrame(
    {'model': model,
     'training': training,
     'test': test
    })
```

```python
df
```

```python
# only grab the model name
df['model'] = df.model.apply(lambda x: str(x).split('(')[0])
```

```python
# form data into tidy format
df = pd.melt(df, id_vars=['model'])
```

```python
df
```

```python
# plot data
plt.subplots(figsize=(7,7))

sns.barplot(x=df.loc[:,'variable'], \
             y=df.loc[:,'value'], \
             hue=df.loc[:,'model'], \
             data=df, alpha=1)
plt.ylim(.675, .82)
```

```python
elasticnet_cv = ElasticNetCV(alphas=alphas, cv=5)

elasticnet_cv.fit(X_train, y_train)

# We are making predictions here
y_preds_train = elasticnet_cv.predict(X_train)
y_preds_test = elasticnet_cv.predict(X_test)

print("Best alpha value is: {}".format(elasticnet_cv.alpha_))
print("R-squared of the model in training set is: {}".format(elasticnet_cv.score(X_train, y_train)))
print("-----Test set statistics-----")
print("R-squared of the model in test set is: {}".format(elasticnet_cv.score(X_test, y_test)))
print("Mean absolute error of the prediction is: {}".format(mean_absolute_error(y_test, y_preds_test)))
print("Mean squared error of the prediction is: {}".format(mse(y_test, y_preds_test)))
print("Root mean squared error of the prediction is: {}".format(rmse(y_test, y_preds_test)))
print("Mean absolute percentage error of the prediction is: {}".format(np.mean(np.abs((y_test - y_preds_test) / y_test)) * 100))
```
