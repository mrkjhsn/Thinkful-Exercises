## LOCATION AND MEAN AGE ANALYSIS

```python
# find top 20 streets/intersections with the most citations 
citations.street.value_counts().head(20).index
```

```python
# subset full citation dataset with top 20 locations
top20_locations = citations[citations['street'].isin(
citations.street.value_counts().head(20).index)]
```

```python
# find locations where the mean age has the most variance over the course of the year
top20_locations.groupby(['month', 'street'])['cited_person_age'].mean().groupby(level=1).apply(lambda x: x.var()).sort_values(ascending=False).head()
```

```python
top5_loc_age_var = citations[citations.street.isin(
    top20_locations.groupby(['month', 'street'])['cited_person_age'] \
    .mean().groupby(level=1).apply(lambda x: x.var()).sort_values(ascending=False).head().index)]
```

```python
top5_loc_age_var.groupby(['month', 'street'])['cited_person_age'].mean().reset_index().head()
```

```python
ax = sns.lineplot(y=top5_loc_age_var.groupby(['month', 'street'])['cited_person_age'].mean().reset_index().iloc[:,2],
            x=top5_loc_age_var.groupby(['month', 'street'])['cited_person_age'].mean().reset_index().iloc[:,0],
            hue=top5_loc_age_var.groupby(['month', 'street'])['cited_person_age'].mean().reset_index().iloc[:,1])


x = top20_locations.groupby(['month'])['cited_person_age'].mean() \
    .reset_index().rename(columns={'cited_person_age': 'Average Locations'}).iloc[:,0]

y1 = top20_locations.groupby(['month'])['cited_person_age'].mean() \
    .reset_index().rename(columns={'cited_person_age': 'Average of Top 20 Locations'}).iloc[:,1]

plt.plot(x, y1, color='black',linewidth=3)
handles, labels = ax.get_legend_handles_labels()

plt.ylabel('Mean Age')
plt.xlabel('Month of Year')
plt.title("Location & Mean Age Analysis Accross the Year")

ax.legend(handles=handles[1:], labels=labels[1:], loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)
```

Mean age at the top 5 locations with citation age variance.  The steepest age related spike happens at E Frank Lloyd Wright Bl/101, during February - the month that the Waste Management Open takes place.
