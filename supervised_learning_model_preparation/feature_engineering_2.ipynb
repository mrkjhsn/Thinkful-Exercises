### 1. Say, we want to understand the relationship between the expenditures of the governments and the students' overall success in the math and reading. Create a new score variable from the weighted averages of all score variables in the datasets. Notice that the number of students in the 4th grade isn't the same as the number of students in the 8th grade. So, you should appropriately weigh the scores!

```python
# find mean scores for reading and math for 4th graders
us_edu["avg_4_all_score"] = us_edu[["avg_math_4_score", "avg_reading_4_score"]].mean(axis=1)

# find mean scores for reading and math for 8th graders
us_edu["avg_8_all_score"] = us_edu[["avg_math_8_score", "avg_reading_8_score"]].mean(axis=1)

# sum number of students in 4th and 8th grades
us_edu["total_4_8_grade"] = us_edu[["grades_4_g", "grades_8_g"]].sum(axis=1)

# find percent of 4th graders that make up total mix of 4th and 8th graders
us_edu["scaled_4_grade"] = us_edu["grades_4_g"] / us_edu["total_4_8_grade"]

# find percent of 8th graders that make up total mix of 4th and 8th graders
us_edu["scaled_8_grade"] = us_edu["grades_8_g"] / us_edu["total_4_8_grade"]

# compute overall score
us_edu["overall_score"] = (us_edu["scaled_4_grade"]*us_edu["avg_4_all_score"] +
                                 us_edu["avg_8_all_score"]*us_edu["scaled_8_grade"])
```

### 3. Now, apply PCA to the 4 expenditure types. How much of the total variance is explained by the 1st component?

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler



X = StandardScaler().fit_transform(us_edu[['instruction_expenditure', 
                                           'support_services_expenditure',
                                           'other_expenditure',
                                           'capital_outlay_expenditure']])

sklearn_pca = PCA(n_components=1)

# what is happening by the creation of this new column?
us_edu["pca_1"] = sklearn_pca.fit_transform(X)

# ratio of total variance in the dataset explained 
# by each component from PCA sklearn
sklearn_pca.explained_variance_ratio_
```
